# Lecture 04 image manipulation

We'll manipulate images $y=mx+b$

```python
from imageio.v3 import imread
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = imread('pisco.jpg')
plt.imshow(img)
```

```python
# Reshape image matrix so that the pixels for each color are in a row
by_color = img.ravel(order = 'F').reshape((3, img.shape[0] * img.shape[1]))

# Create a color intensity histogram
colors = ['red', 'green', 'blue']
_ = plt.hist(by_color.T, 30, histtype='step', color=colors)
```

```python
# Half of red intensity
img_half_red = img.copy()
img_half_red[:, :, 0] = img_half_red[:, :, 0] // 2
plt.imshow(img_half_red)
```

```python
# Color intensity histogram for half of red intensity

by_color = img_half_red.ravel(order = 'F').reshape((3,img_half_red.shape[0] * img_half_red.shape[1]))
_ = plt.hist(by_color.T, 30, density=True, histtype='step', color=colors, label=colors)
```

```python
# Remove blue
img_no_blue = img.copy()
img_no_blue[:, :, 2] = 0
plt.imshow(img_no_blue)
```

```python
# Grayscale image
img_bw = np.uint8(0.2126 * img[:,:,0] + 0.7152 * img[:,:,1] + 0.0722 * img[:,:,2])
plt.imshow(img_bw)
plt.gray()
```
