We'll manipulate images $y=mx+b$

```python
from imageio.v3 import imread
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from mymodule import intensity_histogram
```

```python
img = imread('pisco.jpg')
plt.imshow(img)
```

```python
intensity_histogram(img)
```

```python
# Half of red intensity
img_half_red = img.copy()
img_half_red[:, :, 0] = img_half_red[:, :, 0] // 2
plt.imshow(img_half_red)
```

```python
intensity_histogram(img_half_red)
```

```python
# Remove blue
img_no_blue = img.copy()
img_no_blue[:, :, 2] = 0
plt.imshow(img_no_blue)
```

```python
# Grayscale image
img_bw = np.uint8(0.2126 * img[:,:,0] + 0.7152 * img[:,:,1] + 0.0722 * img[:,:,2])
plt.imshow(img_bw)
plt.gray()
```

```python
w = np.ones([15, 15]) 
img_sharp = signal.fftconvolve(img_bw, w, mode = 'valid')
plt.imshow(img_sharp)
```

```python
# Crop
x, y, _ = img.shape

crop = img[x//4: - x//8, y//4: - y//8, :]

plt.imshow(crop)
```

```python
# Rotate
from scipy import ndimage

rotated = ndimage.rotate(crop, 45)

plt.imshow(rotated)
```
